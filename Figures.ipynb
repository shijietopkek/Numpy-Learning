### Figures and Subplots
- figure - container thats holds all elements of plot(s)
- subplot - appears within a rectangular grid within a figure

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_figure = plt.figure("My Figure!")
```

```python
subplot_1 = my_figure.add_subplot(2,3,1)
subplot_6 = my_figure.add_subplot(2,3,6)
```

```python
plt.plot(np.random.rand(50).cumsum(), 'k--') #dataset, line type= black dashed lines
plt.show()
```

```python
subplot_2 = my_figure.add_subplot(2,3,2)
plt.plot(np.random.rand(50), 'go') #random values in a given shape between [0,1)
plt.show()
```

### different types of graph markers

```python
plt.plot(np.random.rand(100), 'g--') #green dashed line
plt.show()
```

```python
plt.plot(np.random.rand(100), "bo") #blue dots
plt.show()
```

```python
plt.plot(np.random.rand(100), 'r^') #red triangle
plt.show()
```

```python
plt.plot(np.random.rand(100), "yD") #yellow diamond
plt.show()
```

```python
plt.plot(np.random.rand(50), "kd") #orange thin diamond
```
